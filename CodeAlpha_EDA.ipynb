```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load a guaranteed dataset directly from seaborn's built-in library
df = sns.load_dataset('tips')

# Let's rename the columns slightly to make it look like a retail transaction dataset
df = df.rename(columns={
    'total_bill': 'Total_Spent',
    'tip': 'Tip_Amount',
    'sex': 'Customer_Gender',
    'smoker': 'Loyalty_Member',
    'day': 'Day_of_Week',
    'time': 'Shift_Type',
    'size': 'Group_Size'
})

# Display the first few rows
print("--- First 5 Rows ---")
print(df.head())

# Check data types and missing values
print("\n--- Data Structure ---")
print(df.info())
```

```python
# --- 1. Summary Statistics ---
print("--- Financial Summary Statistics ---")
print(df.describe())

# --- 2. Visualizing Spending Distribution ---
plt.figure(figsize=(8, 5))
sns.histplot(df['Total_Spent'], kde=True, color='purple')
plt.title('Distribution of Customer Total Spend')
plt.xlabel('Total Bill Amount ($)')
plt.ylabel('Number of Transactions')
plt.show()

# --- 3. Spending Habits: Loyalty Members vs Non-Members ---
plt.figure(figsize=(8, 5))
sns.boxplot(x='Loyalty_Member', y='Total_Spent', data=df, palette='Set2')
plt.title('Total Spend: Loyalty Members vs Non-Members')
plt.xlabel('Is a Loyalty Member?')
plt.ylabel('Total Spent ($)')
plt.show()
```

```python
# --- 4. Scatter Plot: Total Spent vs Tip Amount ---
plt.figure(figsize=(8, 5))
sns.scatterplot(x='Total_Spent', y='Tip_Amount', hue='Shift_Type', data=df, palette='Dark2')
plt.title('Relationship Between Total Spent and Tip Amount')
plt.xlabel('Total Bill Amount ($)')
plt.ylabel('Tip Amount ($)')
plt.show()
```
